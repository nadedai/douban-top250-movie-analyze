## 三、数据分析
### 1、电影上映年份分析

```python
yearCount = df["年份"].value_counts()
data=yearCount.sort_index() 
print("数量最多的的年份",yearCount.index[0],"其数量为:",df["年份"].value_counts().max())
print("最大的年份",df["年份"].max())
print("最小的年份",df["年份"].min())
```

### 2、每年电影数量变化

```python
plt.figure(figsize=(15,5))
plt.rcParams['font.size'] = 10
plt.xticks(data.index.values,rotation = 90)
plt.yticks(range(df["年份"].value_counts().min(),df["年份"].value_counts().max()+1))
plt.plot(data.index.values,data.values)
plt.xlabel('年份')
plt.ylabel('电影数量')
plt.title('电影数量变化')
```

### 3、上榜次数最多的几个年份

```python
 year = df["年份"].reset_index().groupby(year["年份"]).size().sort_values(ascending = False)
```

```python
x = year.index[:10].astype("str")
y = year.values[:10]
plt.figure(figsize=(7,7),dpi=100)
plt.title("上榜次数最多的几个年份")
plt.ylabel('总计数量')
plt.xlabel('年份')
plt.bar(x,y ,width = 0.5,color='cornflowerblue')
# 添加数值显示
for x_, y_ in zip(x, y):
    plt.text(x_, y_, '{:.2f}'.format(y_), ha='center', va='bottom')
```

### 2、电影制作国家分析  


#### (1) 参数的处理---转换为列表格式
有些电影由多个国家或地区联合制作的。  

```python
country = df["国家"].str.split(' ')
country.head(5)
```

#### (2)参数的处理---转换为Pandas Series格式

```python
country = df["国家"].str.split(' ',expand = True).apply(pd.Series)
country.head()
```

我们可以看到，有些电影甚至有5个国家或地区参与制作。

#### (3)空白值填补
上面的DataFrame有很多空值，对于这么多的空值，可以通过先按列计数，将空值 NaN 替换为“0”，再按行汇总。最后我们统计每个区域里相同国家的总数。

```python
all_country = country.apply(pd.value_counts).fillna(0)
all_country.head()
```

#### (4)总数统计
接下来我们可以计算每个国家参与制作电影总数排名情况：

```python
all_country.columns = ['area1','area2','area3','area4','area5']
all_country['area1'] = all_country['area1'].astype(int)
all_country['area2'] = all_country['area2'].astype(int)
all_country['area3'] = all_country['area3'].astype(int)
all_country['area4'] = all_country['area4'].astype(int)
all_country['area5'] = all_country['area5'].astype(int)
all_country.head()
```

```python
all_country['all_counts'] = all_country['area1']+all_country['area2']+all_country['area3']+all_country['area4']+all_country['area5']
#降序
all_country.sort_values(['all_counts'],inplace = True,ascending=False)
all_country.head(5)
```

由于很多国家只有一部作品 则将他们归入其他类中便于画出图

```python
country = all_country[all_country['all_counts'] <= 5] 
all_country = all_country.drop(country.index).append(pd.Series({"all_counts":country["all_counts"].sum()},name="其他"))
```

####   (3)画出国家占比图

```python
plt.figure(figsize=(10,10))
plt.rcParams['font.size'] = 15
plt.pie(all_country['all_counts'].values,labels = all_country.index,autopct='%1.1f%%',labeldistance = 1,startangle = 90,pctdistance =0.7)
plt.title('国家占比饼图')
```

### 3、电影类型分析
#### (1) 参数的处理---转换为列表格式

```python
all_type = df['类型'].str.split(' ').apply(pd.Series)
all_type.head(10)
```

#### (2)总数统计

```python
all_type = df['类型'].str.split(' ').apply(pd.Series)
all_type = all_type.apply(pd.value_counts).fillna('0')
all_type.columns = ['tpye1','type2','type3','type4','type5']
all_type['tpye1'] = all_type['tpye1'].astype(int)
all_type['type2'] = all_type['type2'].astype(int)
all_type['type3'] = all_type['type3'].astype(int)
all_type['type4'] = all_type['type4'].astype(int)
all_type['type5'] = all_type['type5'].astype(int)
all_type.head(10)

all_type['all_counts'] = all_type['tpye1']+all_type['type2']+all_type['type3']+all_type['type4']+all_type['type5']

all_type = all_type.sort_values(['all_counts'],ascending=False)
all_type.head()
```

由于很多类型只有少数部作品 则将它们归入其他类中便于画出图

```python
types = all_type[all_type['all_counts'] <= 6] 
all_type = all_type.drop(types.index).append(pd.Series({"all_counts":types["all_counts"].sum()},name="其他"))
```

####   (3)画出类型占比图

```python
plt.figure(figsize=(10,10))
plt.rcParams['font.size'] = 15
plt.pie(all_type['all_counts'].values,labels = all_type.index,autopct='%1.1f%%',labeldistance = 1,startangle = 90,pctdistance =0.8)
plt.title('类型占比饼图')
```

### 4、上榜次数最多的导演

Series转DataFrame可以使用字典的方式。

```python
director = df['导演'].value_counts() #value_counts()返回一个Series 序列
director.head()
```

```python
# director.index 可以查看下标 
# director.values 可以查看值
myDirector = pd.DataFrame({'name':director.index,'counts':director.values}) 
myDirector.head()
```

#### 上榜次数最多的几个导演

```python
plt.figure(figsize=(15,5))
plt.rcParams['font.size'] = 15
# [:10][::-1] 取10个 且为倒序
plt.barh(myDirector['name'][:10][::-1],myDirector['counts'].values[:10][::-1])
plt.title("上榜次数最多的几个导演")
plt.xlabel("数量")
```

### 5、评分和排名的关系可视化

```python
plt.figure(figsize=(10,5))
plt.scatter(df['评分'],df['0'])
plt.xlabel('movie_score')
plt.ylabel('movie rank')
#修改y轴为倒序
plt.gca().invert_yaxis()
```

```python
# 集中趋势的直方图
plt.hist(df['评分'],bins=14)
```

```python
# 电影排名和评分的相关性检测 
# 值范围-1到+1，0表示两个变量不相关，正值表示正相关，负值表示负相关，值越大相关性越强
df['评分'].corr(df['0'])
```

评分越高 排名越高(此为负相关 评分越高 值越低 排名是按值从小到大的) 可见豆瓣的评分与电影排名相关的很高
